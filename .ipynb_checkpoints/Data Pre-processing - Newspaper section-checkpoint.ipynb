# Segement each yearly file to seperate documents

```python
import re

doc_index = 1

for i in range(1990, 2013):
    file = open("D:\\coca and GloWbE\\coca\\texts\\text_newspaper_lsp\\w_news_%d.txt" % i)
    newtext = file.read()
    file.close()
    
    p = re.split('\n##\d+', newtext)
    p = p[1:] # due to list.split() the 1st item is blank
    print("File w_news_%d" % i, "-", "Number of files : ", len(p))

    # Using RegEx & Storing each article as a file
    
    for f in p:
        written = open(("D:\\COCA & GloWbE\\COCA\\Pre-processed Text\\Doc2Vec Docs - News\\file_%d.txt" % doc_index), 'w')
        written.write(f)

        doc_index+= 1
        
print("Total number of files:", doc_index)
```

### Newspaper excel sheet 

```python
news_sheet = workbook.sheets()[5] # get sheet 6

# Headers - column "Section" is at 8th col 
print(news_sheet.row_values(0))
```

- Our column of interest 'Section' is the 7th column. <br>
- We iterate through each row and check the 7th column for science related topics <br>
- The section names are rather noisy, so we extract them first to see if we are on the right track

```python
sci_sections = []

for i in range(news_sheet.nrows):
    
    row = news_sheet.row_values(i)
    
    if(("sci" in row[7].lower())): # row[6] == 'section'
        sci_sections.append(row[7])
#    print(mag_sheet.col_values(i)) # mag_sheet.col_values()
```

```python
len(sci_sections)
```

Now let us get the rows with the matching sections

```python
sci_text_rows = []

for i in range(news_sheet.nrows):
    
    row = news_sheet.row_values(i)
    
    if(("sci" in row[7].lower())): # row[6] == 'section'
        sci_text_rows.append(row)
```

```python
sci_text_rows
```

Check if any rows have different lengths with each other

```python
temp = [str(i) for i in sci_text_rows[0]]
temp = '\n'.join(temp)
print("A normal row would have 9 elements, as: \n\n", temp)
```

```python
for k in range(len(sci_text_rows)):    
    for i in range(len(sci_text_rows[k])):

        if i == 0: 
            temp = len(sci_text_rows[i])
            pass

        print(temp == len(sci_text_rows[i]))

        temp = len(sci_text_rows[i])
```

### Now to write out a new excel file with only the relevant entries

```python
import xlsxwriter
processed_workbook = xlsxwriter.Workbook(r'D:\COCA & GloWbE\COCA\COCA Newspaper Science texts')
worksheet = processed_workbook.add_worksheet()
```

```python
original_sheet_COLS = len(news_sheet.row_values(0))
target_sheet_ROWS = len(sci_text_rows)
target_sheet_COLS = len(sci_text_rows[0])

# Write header
for col in range(original_sheet_COLS):
    worksheet.write(0, col, news_sheet.row_values(0)[col])
    
# Write each row by iterating column-wise

for row in range(target_sheet_ROWS):
    for col in range(target_sheet_COLS):
        worksheet.write(row+1, col, sci_text_rows[row][col]) # row+1 because of the header
    
    
```

```python
processed_workbook.close()
```
