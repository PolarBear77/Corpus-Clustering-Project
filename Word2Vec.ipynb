### Parsing all text 

```python
import os

directory = r"C:\nlp\Science-related texts"
file_list = [file for file in os.listdir(directory) if file.endswith('.txt')]

print( len(file_list), "text files detected!")
```

```python
all_text = []

progress = [i * 1000 for i in range(1, 11) ] 
k = 0

for i, text in enumerate(file_list):
    
    if(i > progress[k]): # display after every thoudsand files
        print("File number", i, "parsed")
        k += 1
        
    try:
        text = preprocess(os.path.join(directory, text))
        text = text.split(" ")[1:-1]  # filter blanks
        text = [word.lower() for word in text]
        text = " ".join(text)
        text = parse_text(text, show_pos = True)
        text = lemmatizer(text, acceptPeriods= True)
        text = removeStopWords(text)
        all_text.extend(text)
    
    except Exception as e: 
        print("Error at file", i, ":", e)
```

```python
import re

class sentenceGen(object):
    """
    Creates a sentence until it meets "._." (period with period tag)
    Upon meeting a period, concatenates the words collected so far and returns them as a sentence
    """

    def __init__(self, text):
        self.text= text.split(" ")
        
    def __iter__(self):
        sentence = []
        
        for word in self.text:
            if word != "._.":
                sentence.append(word)
            elif word == "._.":
                yield sentence
                sentence = []
```

```python
import gensim
import os
import logging

# Log settings
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)


# a memory - friendly iterator
# sentences = MySentences('D:\\coca and GloWbE\\coca\\texts\\text_newspaper_lsp')
model = gensim.models.Word2Vec(sentenceGen(all_text))
```
